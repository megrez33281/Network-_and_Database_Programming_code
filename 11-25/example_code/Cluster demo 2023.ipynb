```
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

features, true_labels = make_blobs(
    n_samples=400, 
    centers=3, 
    cluster_std=2.75, 
    random_state=42
)
```

```
features[:5]
```

```
true_labels[:5]
```

```
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
#將資料進行標準化
```

```
scaled_features[:5]
```

```
plt.figure()
plt.scatter(scaled_features[:,0], scaled_features[:,1],s=100, cmap='voridis')
plt.title('Dataset')
plt.show()
```

kmeans = KMeans(
    init="random", n_clusters=3,
    n_init=10, max_iter=300,
    random_state=42
)
kmeans.fit(scaled_features)

```
kmeans.cluster_centers_
```

```
kmeans.n_iter_
```

```
kmeans.labels_
```

```
kmeans = KMeans(
    init="random", n_clusters=3,
    n_init=10, max_iter=300,
    random_state=42
)
kmeans.fit(scaled_features)

plt.figure()
plt.scatter(scaled_features[:,0], scaled_features[:,1], c=kmeans.labels_, s=100)
plt.title('Dataset')
plt.show()
```
